## Data Preprocessing

```python
data1=data.copy()
```

```python
## Scale the data
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
data1[['Age', 'Gender', 'Technology_Usage_Hours', 'Social_Media_Usage_Hours',
       'Gaming_Hours', 'Screen_Time_Hours',
       'Stress_Level', 'Sleep_Hours', 'Physical_Activity_Hours',
       'Support_Systems_Access', 'Work_Environment_Impact',
       'Online_Support_Usage']] = scale.fit_transform(data1[['Age', 'Gender', 'Technology_Usage_Hours', 'Social_Media_Usage_Hours',
       'Gaming_Hours', 'Screen_Time_Hours',
       'Stress_Level', 'Sleep_Hours', 'Physical_Activity_Hours',
       'Support_Systems_Access', 'Work_Environment_Impact',
       'Online_Support_Usage']])
```

```python
data1.head()
```

## Create X and y

```python
X = data1.drop('Mental_Health_Status',axis=1)
y = data1.Mental_Health_Status
```

```python
X.head(3)
```

```python
y.head(3)
```

```python
## split the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,
                                                 random_state=18)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
```

```python
smote = SMOTE()
X_smote,y_smote = smote.fit_resample(X_train,y_train)
```

```python
print('Actual value',Counter(y_train))
print('SMOTE value',Counter(y_smote))
```

# Create Models

## Model 1: Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_smote,y_smote)
dt_pred = dt.predict(X_test)

```

```python
## Check Metrics 
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_test,dt_pred))
```

```python
print(classification_report(y_test,dt_pred))
```

## Model 2: Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_smote,y_smote)
rf_pred = rf.predict(X_test)
```

```python
accuracy_score(y_test,rf_pred)
```

```python
print(classification_report(y_test,rf_pred))
```

```python
pd.crosstab(y_test,rf_pred)
```

## Model 3: Random Forest Classifier with RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV

params = {
    'n_estimators':[50,100,150,200,250,300],
    'max_depth':(list(range(1,20))),
    'max_features':['auto','sqrt','log2'],
    'min_samples_split':[2,3,4],
    'min_samples_leaf':(list(range(1,20))),
    'bootstrap':[True,False]
}

rf2 = RandomForestClassifier()
rf_cv = RandomizedSearchCV(estimator=rf2, param_distributions=params, cv = 3, verbose = 2, scoring='accuracy', n_jobs=-1, n_iter=100)

rf_cv.fit(X_smote,y_smote)
best_param = rf_cv.best_params_
print(f'Best Parameters:{best_param}')
```

```python
rf2 = RandomForestClassifier(n_estimators=100,min_samples_split=2,min_samples_leaf=3,max_features='sqrt',max_depth=21,bootstrap=False)
rf2.fit(X_smote,y_smote)
rf2_pred = rf2.predict(X_test)
```

```python
accuracy_score(y_test,rf2_pred)
```

```python
print(classification_report(y_test,rf2_pred))
```

## Model Comparison Summary:

You have compared the performance of several machine learning models based on their accuracy. Hereâ€™s a summary of the results:

1. **Random Forest** and **Bagging Classifier** models achieved the highest accuracy of **85.5%**. These models are ensemble methods, which often perform well in classification tasks by combining the predictions of multiple base models.

2. **Random Forest Classifier with RandomizedSearchCV** slightly lags behind with an accuracy of **85.3%**. This approach involves hyperparameter tuning, which can sometimes lead to slight variations in accuracy but can also improve model performance if done correctly.

3. **Decision Tree** is the simplest model among the ones compared, and it achieved an accuracy of **84.6%**. Decision Trees are prone to overfitting, especially without pruning or when not used in an ensemble method like Random Forest.

4. **XGB Boost** (XGBoost) has an accuracy of **84.2%**. XGBoost is a powerful boosting algorithm known for its high performance, but in this case, it did not outperform the ensemble methods.

5. **XGB with RandomizedSearchCV** showed a slight improvement over standard XGBoost, achieving **84.4%** accuracy. Like the Random Forest, this model benefits from hyperparameter tuning.

### Key Takeaways:
- **Ensemble methods** (Random Forest and Bagging Classifier) are the top performers, suggesting that combining multiple models or trees can lead to better accuracy.
- **Hyperparameter tuning** (as done with RandomizedSearchCV) can fine-tune models for better performance, though the gains may be incremental.
- While **Decision Tree** and **XGBoost** are strong contenders, they didnâ€™t outperform the ensemble methods in this particular dataset.

If youâ€™re looking to optimize further, focusing on hyperparameter tuning or exploring other ensemble methods might provide additional gains in accuracy.
